```python
import pandas as pd
```

Reading in the list of prepositions and committees from CSV.

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
props.info()
```

```python
props.head()
```

Reading in the list of contribuitions to ballot mesures

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

```python
contribs.head()
```

```python
contribs.info()
```

```python
props.prop_name.value_counts().reset_index()
```

```python
props.prop_name.value_counts()
```

```python
prop = props[props.prop_name == "PROPOSITION 063- FIREARMS. AMMUNITION SALES. INTIATIVE STATUTE."]
```

```python
prop.info()
```

```python
merged = pd.merge(prop, contribs, on = "calaccess_committee_id")
```

```python
merged.info()
```

```python
merged.amount.sum()
```

```python
merged.committee_position.value_counts()
```

```python
support = merged[merged.committee_position == "SUPPORT"]
```

```python
oppose = merged[merged.committee_position == "OPPOSE"]
```

```python
support.amount.sum()
```

```python
oppose.amount.sum()
```

```python
support.amount.sum() / merged.amount.sum()
```

```python
oppose.amount.sum() / merged.amount.sum()
```

```python
merged.sort_values("amount", ascending = False)
```

```python
oppose.sort_values("amount", ascending = False).head(100)
```

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending = False)
```

```python
merged.fillna("no_name").groupby(["contributor_firstname", "contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending = False).info()
```

```python
oppose.fillna("no_name").groupby(["contributor_firstname", "contributor_lastname", "committee_position", "committee_name_x"]).amount.sum().reset_index().sort_values("amount", ascending = False)
```
